Handling the Missing Values

step 1:convert all -200 to NAN

step 2:replace all NAN values with mean of that specific column

```python
air_quality_data=air_quality_data.replace(to_replace=-200,value=np.NAN)
```

```python
# for checking the number if missing values
air_quality_data.isnull().sum()
```

This show the actual number of missing values

```python

air_quality_data.tail()
```

```python
air_quality_data.select_dtypes(include='number').mean()
```

```python
# replacing the missing values with mean value

air_quality_data=air_quality_data.fillna(air_quality_data.mean())
```

```python
# Select only numeric columns
numeric_cols = air_quality_data.select_dtypes(include='number').columns

# Replace missing values in numeric columns with the mean of that column
air_quality_data[numeric_cols] = air_quality_data[numeric_cols].fillna(air_quality_data[numeric_cols].mean())
```

```python
# converting the date from DD/MM/YYYY TO YYYY-MM-DD
data_info= pd.to_datetime(air_quality_data['Date'])
print(data_info)
```

```python
date_info= pd.to_datetime(air_quality_data['Date'], dayfirst=True)
print(date_info)
```

```python
time_info=air_quality_data["Time"]
```

```python
time_info=time_info.apply(lambda x:x.replace('.',':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining 2 series to a pandas dataframes
date_time=pd.concat([date_info,time_info],axis=1)
```

```python
print(date_time)
```

```python
# combine  date and time in column

date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
```

```python
# convert ds from object datatype to datetime format
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
# traning the prophet model
m=Prophet()
m.fit(data)
```

```python
# future prediction
future =m.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_lower']].tail()
```

```python
# plot
m.plot(forecast)
```

```python
# plot
m.plot_components(forecast)
```
